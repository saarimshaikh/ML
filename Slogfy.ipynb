```python
import csv, smtplib
from email.message import EmailMessage
server = smtplib.SMTP("smtp.gmail.com", 587)
server.starttls()
server.login('saarimshaikh21@gmail.com', 'password')
subject="This is a demo mail for my internship"
msg = '''Hello\n
This is my demo mail for my internship thing.\n

Regards\n
Slogfy'''
sender='saarimshaikh21@gmail.com'

with open("latest.csv") as file:
  reader=csv.reader(file)
  next(reader)
  for addr in reader:
    email=EmailMessage()
    email['From']=sender
    email['To']=addr
    email['Subject']=subject
    email.set_content(msg)
    server.send_message(email)
    print(f'Sent to {addr} ')
server.close()
```
